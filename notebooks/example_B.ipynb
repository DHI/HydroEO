This notebook outlines a more detailed process for setting up an dowloading data for a simple altimetry project. It is intended for those who wish to have more controll over the individual steps in the processing of data.

Data for three example reservoirs in the mekong river basin is used to outline the basic steps necessary to get data simply and quickly.

This example uses the configuration file named "config_B.yaml". You will need to edit the lines marked "EDIT" to ensure that the notebook runs properly. If you have not looked at example_A.ipynb, it is encouraged to do so so that you can understand the basic flow and setup for an altimetry project.

```python
# First lets load the needed packages
import yaml
from altimetry.project import Project
```

First lets inspect the reservoirs and their location for where we would like to download data. We can read in information from the config file or look directly to inspect the reservoirs. Make sure that the path is correct.

```python
# path_to_config = r"..\notebooks\config_A.yaml"
path_to_config = r"..\data\clark_hill_config.yaml"
with open(path_to_config, "rt") as f:
    config = yaml.safe_load(f.read())

path_to_reservoirs = config["reservoirs"]["path"]
print(f"Path to reservoirs: {path_to_reservoirs}")
```

Now that we are sure we are referencing the right reservoirs for our project we can go ahead and create the altimetry project. We will give it a meaningful name and point to the configuration file.

```python
mekong = Project(name="Mekong", config=path_to_config)
mekong.report()
```

The reservoirs object can be extracted from the alitmetry project object in order to run specific functions yourself.

```python
# Here we extract the reservoirs class from the altimetry project in order to investigate the functions in more detail
reservoirs = mekong.reservoirs
reservoirs.report()

# We can view an interactive map of the reservoirs by inspecting the geopandas geodataframe attribute
reservoirs.gdf.explore()
```

Here we run some of the functions needed before downloading data. If downloading SWOT data this step will make sure that the PLD (Prior Lake Database) for the area surrounding the reservoirs is downloaded or properly referenced. The PLD is the reference database used by the SWOT to associate observations to a known waterbody, allowing for a quick reliable water surface elevation timeseries. This code block is represented within the project.initialize() function.

```python
# download PLD from hydroweb if needed, the overwrite argument can be provide3d to force a redownload
reservoirs.download_pld(overwrite=False)

# take the downloaded PLD and see where we have overlap with the input reservoirs,
# we can set a maximum distance threshold when looking for overlap between the pld.
# This should typically be quite small. Max_distance is in meters
reservoirs.assign_pld_id(local_crs=mekong.local_crs, max_distance=100)

# flag and export which reservoirs have entries in the PLD
reservoirs.flag_missing_priors()

# assign download geometry (for reservoirs this is the same as the input shape)
# Here you could edit some of your reservoir polygons before setting the download gdf in order to refine where altimetry crossings are allowed
reservoirs.download_gdf = reservoirs.gdf
```

We are now ready to download data for reservoir. Depending on how many reservoirs and products you are downloading as well as the date range, downloads may take a while. If a download stops, you should be able to rerun the command without duplicating downloads. You may iterate here until you have sucessfully downloaded your altimetry data.

While the products listed for download are defined in the configuration file, you can also select which products to run downloads for and there individual download dates here.

```python
reservoirs.download_altimetry(
    product="SWOT_Lake",
    startdate=mekong.startdates["swot"],
    enddate=mekong.enddates["swot"],
    update_existing=False,  # argument to determine if to use the end date or download data from start date up to current date
)
reservoirs.download_altimetry(
    product="ATL13",
    startdate=mekong.startdates["icesat2"],
    enddate=mekong.enddates["icesat2"],
    update_existing=False,
)
reservoirs.download_altimetry(
    product="S3",
    startdate=mekong.startdates["sentinel3"],
    enddate=mekong.enddates["sentinel3"],
    credentials=(mekong.creodias_user, mekong.creodias_pass),
    update_existing=False,
)
reservoirs.download_altimetry(
    product="S6",
    startdate=mekong.startdates["sentinel6"],
    enddate=mekong.enddates["sentinel6"],
    credentials=(mekong.creodias_user, mekong.creodias_pass),
    update_existing=False,
)
```

The functions for extracting valid observations, cleaning the timeseries, and merging different products into a single timeseries are shown in the next few code blocks. These functions are implemented within the project.create_timeseries() function. Future development could include more options for personalized processing parameters within these functions.


Extraction of data: The function below takes the raw data and combines it into a single shape file holding observations and their locations. You may provide a list of the products that you wish to process. The product options are shown below.

```python
print(mekong.to_process)
reservoirs.extract_product_timeseries(mekong.to_process)
```

You can clean the raw data by product before merging to remove any obvious outliers. Currently you must provide a list of strings that correspond to the filters. To see which filters are available look at the function source code. The elevation and MAD filters are the two default filters within the project workflow.

```python
reservoirs.clean_product_timeseries(
    products=mekong.to_process, filters=["elevation", "MAD"]
)
```

After cleaning individual timeseries, we merge them into a final timeseres. with the following function.

```python
reservoirs.merge_product_timeseries(products=mekong.to_process)
```

The following points out the functions called by the project.generate_summaries() function. You can decide whether you wish to save and/or show the figures with the show and save arguments. These figures are generated directly from the saved timeseries. Investigating the steps can help you decide which aspects of processing and cleaning may be leading to issues in your workflow. Future development could aim to allow specific processing steps to be omited for a given reservoir.

```python
for id in reservoirs.download_gdf[reservoirs.id_key]:
    reservoirs.summarize_crossings_by_id(id, show=True, save=False)
    reservoirs.summarize_cleaning_by_id(id, show=True, save=False)
    reservoirs.summarize_merging_by_id(id, show=True, save=False)
```

These steps run a single set of processing steps and parameters for each reservoir. Each reservoir will have different available data and lenght of timeseries, meaning that different processing steps may be better for some reservoirs rather than others.
