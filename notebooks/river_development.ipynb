This notebook shows how to download ICESat-2 (ATL13 - Inland Water Surface Height) data for a given area of interest.

```python
import os
import matplotlib.pyplot as plt
import shapely
import pandas as pd
import geopandas as gpd
from datetime import date

from altimetry import icesat2, utils, geometry
```

Load in area of interest

```python
# example looking at various rivers
mekong_rivers = gpd.read_file(r"..\gis\main-stream.zip")
print(f"Number of rivers: {len(mekong_rivers)}")
mekong_rivers.head()
```

```python
# standardize all inputs to the same global crs and estimate the local crs
global_crs = 'EPSG:4326'
local_crs = mekong_rivers.estimate_utm_crs()

# buffer rivers within local crs
buffer_width = 1000 # meters
mekong_rivers_buffered = mekong_rivers.copy()
mekong_rivers_buffered.geometry = mekong_rivers.to_crs(local_crs).buffer(buffer_width)

# push everything to global crs
mekong_rivers = mekong_rivers.to_crs(global_crs)
mekong_rivers_buffered = mekong_rivers_buffered.to_crs(global_crs)

# break down large river system into grid for download and search
area_bounds = mekong_rivers_buffered.unary_union.bounds
grid_gdf = gpd.GeoDataFrame(geometry=geometry.grid_bounds(area_bounds, 0.5), crs=mekong_rivers_buffered.crs)
grid_gdf = grid_gdf.loc[grid_gdf.intersects(mekong_rivers_buffered.unary_union)].reset_index(drop=True)
grid_gdf['id'] = grid_gdf.index

# plot results
fig, main_ax = plt.subplots(1, 2)

ax = main_ax[0]
mekong_rivers_buffered.plot(ax=ax, color='red')
mekong_rivers.plot(ax=ax)
grid_gdf.plot(ax=ax, edgecolor='black', facecolor='None')
ax.set_title('Full River System')

ax = main_ax[1]

river_index = 2
mekong_rivers_buffered.loc[[river_index]].plot(ax=ax, color='red')
mekong_rivers.loc[[river_index]].plot(ax=ax)
ax.set_title(f"River: {mekong_rivers.loc[river_index, 'riv_name']}")
fig.tight_layout()
plt.show()
```

```python
for i in grid_gdf.index[62:]:

    # grab coordinates of river
    coords = [(x, y) for x, y in grid_gdf.loc[i, 'geometry'].exterior.coords]
    id = grid_gdf.loc[i, 'id']

    # define bounds of data search
    startdate = date(2019, 1, 1)
    enddate   = date(2024, 10, 15)

    # define and if needed create directory for each river stretch
    download_directory=rf'..\data\icesat2\{id}'
    utils.ifnotmakedirs(download_directory)

    # make a simple aoi based on coordinates
    order_ids = icesat2.query(aoi=coords, startdate=startdate, enddate=enddate, earthdata_credentials=None, download_directory=download_directory, product='ATL13')

    # check that we downloaded some files
    print("Downloaded files:")
    for f in os.listdir(download_directory): print(f"\t- {f}") 
```

Quickly inspect a download grid cell

```python
from matplotlib.dates import date2num
from datetime import datetime

i = 46
download_directory=rf'..\data\icesat2\{i}'
files = list(os.listdir(download_directory))

fig, main_ax = plt.subplots()
ax = main_ax

# set boudns
xmin, ymin, xmax, ymax = grid_gdf.loc[i, 'geometry'].bounds
ax.set_xlim([xmin-0.1, xmax+0.1])
ax.set_ylim([ymin-0.1, ymax+0.1])

# plot rid and river
grid_gdf.loc[[i]].plot(ax=ax, edgecolor='black', facecolor='None')
#mekong_rivers_buffered.plot(ax=ax, color='red')
mekong_rivers.plot(ax=ax)

# load and plot all files for all tracks and crossings
for file in files:
    print(file)
    infile= os.path.join(download_directory, file)
    for key in ['gt1l', 'gt1r', 'gt2l', 'gt2r', 'gt3l', 'gt3r']:

        data = icesat2.ATL13(infile, key)

        if data.check_height_data():
            data_df = data.read()
            data_gdf = gpd.GeoDataFrame(data_df, geometry=gpd.points_from_xy(data_df.lon, data_df.lat))
            data_gdf = data_gdf.loc[data_gdf.within(mekong_rivers_buffered.unary_union)].reset_index(drop=True)

            if len(data_gdf) > 0:
                data_gdf['color'] = [int(date2num(i)) for i in data_gdf["date"].values]
                data_gdf.plot(ax=ax, column='color', vmin=int(date2num(datetime(2019, 1, 1))), vmax=int(date2num(datetime(2024, 12, 31))), cmap='jet', alpha=0.5)

fig.tight_layout()
plt.show()
```

Inspect all data downloaded along river

```python
center_points = list()

for i in grid_gdf.index:
    print(f"{i}/{len(grid_gdf)}")
    download_directory=rf'..\data\icesat2\{i}'

    if os.path.exists(download_directory):

        # load and plot all files for all tracks and crossings
        files = list(os.listdir(download_directory))        
        for file in files:
            infile= os.path.join(download_directory, file)

            for key in ['gt1l', 'gt1r', 'gt2l', 'gt2r', 'gt3l', 'gt3r']:

                data = icesat2.ATL13(infile, key)
                
                if data.check_height_data():
                    data_df = data.read()
                    data_gdf = gpd.GeoDataFrame(data_df, geometry=gpd.points_from_xy(data_df.lon, data_df.lat))
                    data_gdf = data_gdf.loc[data_gdf.within(mekong_rivers_buffered.unary_union)].reset_index(drop=True)

                    if len(data_gdf) > 0:
                        center_points.append(data_gdf.loc[[0]])
                        
center_points = pd.concat(center_points)
```

```python
fig, main_ax = plt.subplots()
ax = main_ax

# plot river
mekong_rivers.plot(ax=ax)
center_points['color'] = [int(date2num(i)) for i in center_points["date"].values]
center_points.plot(ax=ax, column="color", vmin=int(date2num(datetime(2019, 1, 1))), vmax=int(date2num(datetime(2024, 12, 31))), cmap='Spectral')

fig.tight_layout()
plt.show()
```
