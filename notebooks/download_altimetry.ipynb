This notebook shows how to download ICESat-2 (ATL13 - Inland Water Surface Height) data for a given area of interest.

```python
import os
import matplotlib.pyplot as plt
import geopandas as gpd
from datetime import date
from altimetry import icesat2
```

Load in area of interest

```python
# example looking at various rivers
mekong_rivers = gpd.read_file(r"..\gis\main-stream.zip")

# standardize all inputs to the same global crs and estimate the local crs
global_crs = 'EPSG:4326'
local_crs = mekong_rivers.estimate_utm_crs()

# buffer rivers within local crs
buffer_width = 1000 # meters
mekong_rivers_buffered = mekong_rivers.copy()
mekong_rivers_buffered.geometry = mekong_rivers.to_crs(local_crs).buffer(buffer_width)

# push everything to global crs
mekong_rivers = mekong_rivers.to_crs(global_crs)
mekong_rivers_buffered = mekong_rivers_buffered.to_crs(global_crs)

# plot results
fig, main_ax = plt.subplots(1, 2)

ax = main_ax[0]
mekong_rivers_buffered.plot(ax=ax, color='red')
mekong_rivers.plot(ax=ax)
ax.set_title('Full River System')

ax = main_ax[1]
mekong_rivers_buffered.loc[[0]].plot(ax=ax, color='red')
mekong_rivers.loc[[0]].plot(ax=ax)
ax.set_title(f"River: {mekong_rivers.loc[1, 'riv_name']}")
fig.tight_layout()
plt.show()
```

```python
# grab coordinates of river
coords = [(x, y) for x, y in mekong_rivers_buffered.loc[1, 'geometry'].exterior.coords]

startdate = date(2019, 1, 1)
enddate   = date(2024, 10, 15)

download_directory=r'..\data\icesat2'

# make a simple aoi based on coordinates
order_ids = icesat2.query(aoi=coords, startdate=startdate, enddate=enddate, earthdata_credentials=None, download_directory=download_directory, product='ATL13')

# check that we downloaded some files
print("Downloaded files:")
for f in os.listdir(download_directory): print(f"\t- {f}") 
```

```python
# inspect one of these files to see where it crosses
infile= os.path.join(download_directory, list(os.listdir(download_directory))[0])
track_keys = ['gt1l', 'gt1r', 'gt2l', 'gt2r', 'gt3l', 'gt3r'] # take all icesat beams


fig, main_ax = plt.subplots()
ax = main_ax
mekong_rivers_buffered.loc[[0]].plot(ax=ax, color='red')
mekong_rivers.loc[[0]].plot(ax=ax)

for key in track_keys:

    try:
        data = icesat2.ATL13(infile, key)
        data_df = data.read()
        data_gdf = gpd.GeoDataFrame(data_df, geometry=gpd.points_from_xy(data_df.lon, data_df.lat))
        data_gdf.plot(ax=ax)
    except:
        print('Data read failed')

fig.tight_layout()
plt.show()


```
